# Iterative Statements

### (1)  For loop

##### syntax:
for x in sequence:
    body

##### 1. WAPP to print characters present in the given string

```python
s="Sahasra"
for x in s:
    print(x)
```

##### 2. To print characters present in string index wise.

```python
s=input("Enter Some string: ")
i=0
for x in s:
    print("The character present at ",i,"index is ",x)
    i+=1
```

```python
s='Hello'
for i in range(1,11):
    print(s)
```

```python
for i in range (10,0,-1):
    print(i)
```

```python
list=eval(input("Enter List: "))
sum = 0
for x in list:
    sum=sum+x
print("The Sum is ",sum)
```

## (2) while Loop

#### If we want to execute a group of statements iteratively until some condition false,then we should go for while
loop.
Syntax:
    while condition:
        body

```python
x=1
while x <=10:
    print(x)
    x=x+1
```

```python
n = int(input("Enter number: "))
sum = 0
i = 1
while i<=n:
    sum+=i
    i+=1
print("The sum of first ", n, "numbers is ", sum)
```

#### 3. WAPP to prompt user to enter some name until entering Karthi

```python
name=""
while name!="Karthi":
    name=input("Enter name: ")
print("Thanks for the confirmation")
```

#### INFINITE LOOPS

```python
i=1
while True:
    print("Hello", i)
    i+=1
    if (i==10000):
        break
```

#### NESTED LOOPS

```python
for i in range(3):
    for j in range(3):
        print("Hello")
```

```python
for i in range(4):
    for j in range(4):
        #print("i= ", i , "j= ", j)
        print('i= {} j = {}'.format(i,j))
```

```python
n = int(input("Enter number of rows: "))
for i in range(1, n+1):
    for j in range(1, i+1):
        print("*", end="")
    print()
```

```python
n=int(input("Enter Number of rows: "))
for i  in range(1, n+1):
    print("* " * i)
```

```python
n=int(input("Enter number of rows: "))
for i in range (1, n+1):
    print(" " * (n), end="") #indent the entire triangle by n (number of rows) spaces
    print("* " * i)
```

```python
n=int(input("Enter number of rows: "))
for i in range (1, n+1):
    print(" " * (n-i), end="") #indent the entire triangle by n (number of rows) spaces
    print("* " * i)
```
